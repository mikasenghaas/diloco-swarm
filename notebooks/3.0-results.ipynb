Nice! Let's also compute this in terms of communication cost.

```python
# Helpers to compute SWARM communication cost
def compute_pp_cost_per_step(num_stages: int, batch_size: int, seq_length: int, hidden_dim: int, bytes_per_param: int = 4) -> float:
    """Compute communication cost for pipeline parallelism in bytes per step
    
    In pipeline parallelism, each stage needs to send its activations forward and gradients backward
    to its adjacent stages. This happens (num_stages-1) times for both forward and backward passes.
    
    Args:
        num_stages: Number of pipeline stages
        batch_size: Micro batch size
        seq_length: Sequence length
        hidden_dim: Hidden dimension size
        bytes_per_param: Number of bytes per parameter (default 4 for float32)
    """
    return 2 * (num_stages-1) * batch_size * seq_length * hidden_dim * bytes_per_param

def compute_dp_cost_per_step(num_stages: int, num_nodes_per_stage: int, num_params:int, bytes_per_param: int = 4) -> float:
    """Compute communication cost for data parallelism in bytes per step
    
    In data parallelism, each group of nodes needs to perform all-reduce on their gradients.
    This happens independently for each pipeline stage.
    
    Args:
        num_stages: Number of pipeline stages
        num_nodes_per_stage: Number of data parallel nodes per stage
        model_size: Total number of model parameters
        bytes_per_param: Number of bytes per parameter (default 4 for float32)
    """
    model_size_per_stage = (num_params / num_stages) * bytes_per_param
    all_reduce_cost = lambda n, x: 2 * (n-1)/n * x
    return num_stages * all_reduce_cost(num_nodes_per_stage, model_size_per_stage)
```

Let's compute communication cost history for different model sizes and methods.

```python
STEPS = 2000
NUM_STAGES = 2
BATCH_SIZE = 2048
SEQ_LENGTH = 1024
NUM_NODES_PER_STAGE = 4
BYTES_PER_PARAM = 4
MODELS = {
    "GPT-2 Small": {"num_params": 180 * 10**6, "hidden_dim": 768}, # 180M
    "GPT-2 Medium": {"num_params": 405 * 10**6, "hidden_dim": 1024}, # 405M
    "GPT-2 Large": {"num_params": 800 * 10**6, "hidden_dim": 1200}, # 800M
    "GPT-2 XL": {"num_params": 1500 * 10**6, "hidden_dim": 1600}, # 1.5B
    "GPT-3": {"num_params": 175 * 10**9, "hidden_dim": 12288} # 175B
}
METHODS = {
    "SWARM": {"sync_every_n_steps": 1},
    "DiLoCo-SWARM": {"sync_every_n_steps": 50}
}

EXPERIMENTS = []
for model_name, model in MODELS.items():
    for method, method_dict in METHODS.items():
        EXPERIMENTS.append({
            "method": method,
            "model": model_name,
            "num_params": model["num_params"],
            "hidden_dim": model["hidden_dim"],
            "sync_every_n_steps": method_dict["sync_every_n_steps"],
            "batch_size": BATCH_SIZE,
            "seq_length": SEQ_LENGTH,
            "num_stages": NUM_STAGES,
            "num_nodes_per_stage": NUM_NODES_PER_STAGE,
            "bytes_per_param": BYTES_PER_PARAM,
            "steps": STEPS
        })
```

```python
import numpy as np
comm_cost = pd.DataFrame()
for exp in EXPERIMENTS:
    steps = np.arange(1, exp["steps"]+1)
    pp_cost_per_step = compute_pp_cost_per_step(
        num_stages=exp["num_stages"], 
        batch_size=exp["batch_size"], 
        seq_length=exp["seq_length"], 
        hidden_dim=exp["hidden_dim"], 
        bytes_per_param=exp["bytes_per_param"]
    )
    dp_cost_per_step = compute_dp_cost_per_step(
        num_stages=exp["num_stages"], 
        num_nodes_per_stage=exp["num_nodes_per_stage"], 
        num_params=exp["num_params"], 
        bytes_per_param=exp["bytes_per_param"]
    )
    dp_costs = np.array([dp_cost_per_step] * (exp["steps"]))
    dp_costs[steps % exp["sync_every_n_steps"] != 0] = 0
    pp_costs = np.array([pp_cost_per_step] * (exp["steps"]))
    swarm_costs = dp_costs + pp_costs
    comm_cost = pd.concat([comm_cost, pd.DataFrame({
        "method": exp["method"],
        "model": exp["model"],
        "dp_cost": np.cumsum(dp_costs),
        "pp_cost": np.cumsum(pp_costs),
        "swarm_cost": np.cumsum(swarm_costs),
        "step": steps
    })])

comm_cost
```

```python
# Merge communication cost with experiment history
exp1_comm = exp1.merge(comm_cost[
    (comm_cost.method.isin(["SWARM", "DiLoCo-SWARM"])) &
    (comm_cost.model == "GPT-2 Small")
    ], left_on=["step", "experiment"], right_on=["step", "method"], how="left")
exp1_comm["swarm_cost"] = exp1_comm["swarm_cost"].fillna(0) / 1e12 # Convert to TB
exp1_comm = exp1_comm.dropna() # Drop Single-GPU (no communication)

sns.set_style("whitegrid")
colors = sns.color_palette("Set2", n_colors=3)[1:]
fig, ax = plt.subplots(figsize=(6, 4), dpi=300)
order = exp1_comm.groupby("experiment")["val_perplexity"].last().sort_values(ascending=False).index
sns.lineplot(data=exp1_comm, x="swarm_cost", y="val_perplexity", hue="experiment", hue_order=order, ax=ax, palette=colors)
ax.set_xlabel("Communication Cost (TB)", fontsize=14); ax.set_ylabel("Perplexity", fontsize=14)
ax.set_ylim((21, 100));
ax.get_xaxis().set_major_locator(plt.MultipleLocator(5))
ax.get_xaxis().set_minor_locator(plt.MultipleLocator(1))
ax.get_yaxis().set_major_locator(plt.MultipleLocator(10))
ax.get_yaxis().set_minor_locator(plt.MultipleLocator(5))
ax.grid(True, which='both', alpha=0.2); ax.grid(True, which='major', alpha=0.5)
ax.legend(title="");

fig.savefig("report/figures/experiment1-2.pdf", bbox_inches="tight")
```

```python
# Show total comm cost (broken down in DP and PP) for increasing model size
fig, ax = plt.subplots(figsize=(8, 3), dpi=300)

total_comm = comm_cost.groupby(["model", "method"])\
    .agg({"dp_cost": "max", "pp_cost": "max", "swarm_cost": "max"})\
    .sort_values("swarm_cost").reset_index()
total_comm

# Melt data (create cost_type, cost columns to group by cost_type)
colors = sns.color_palette("Set2", n_colors=3)[1:][::-1]
total_comm_melted = total_comm.melt(
    id_vars=["model", "method"], 
    value_vars=["dp_cost", "pp_cost", "swarm_cost"],
    var_name="cost_type",
    value_name="cost"
)

# Filter and pivot data for stacked bar plot
plot_data = total_comm_melted[
    (total_comm_melted.cost_type != "swarm_cost") & 
    (total_comm_melted.method == "SWARM")
].pivot(index="model", columns="cost_type", values="cost")

order = ["GPT-2 Small", "GPT-2 Medium", "GPT-2 Large", "GPT-2 XL", "GPT-3"]
plot_data = plot_data.reindex(order)
plot_data /= 1e12  # Convert bytes to terabytes

# Create stacked bar plot
ax = plot_data.plot(kind='bar', stacked=True, ax=ax, color=colors)

# Add percentage labels inside bars
total = plot_data.sum(axis=1)
dp_pct = plot_data['dp_cost'] / total * 100

for i, (pct, total_val) in enumerate(zip(dp_pct, total)):
    # Position text at top of bar
    y_pos = total_val
    ax.text(i, y_pos, f'{pct:.1f}%', ha='center', va='bottom', fontsize=14)

# Fade the last bar slightly
alpha = 0.7; ax.patches[-6].set_alpha(alpha); ax.patches[-1].set_alpha(alpha)

# Rotate x-axis labels to be horizontal  
plt.xticks(rotation=0)

ax.set_xlabel("Model", fontsize=14)
ax.set_ylabel("Communication Cost (TB)", fontsize=14)
ax.legend(title="Cost Type", labels=["DP", "PP"], fontsize=14);
ax.set_yscale('log'); ax.set_ylim(top=ax.get_ylim()[1]*1.5)

fig.savefig("report/figures/square-cube-law.pdf", bbox_inches="tight")
```

## Benchmark

In this experiment, we are benchmarking the throughput of the four types of parallelism implemented:

* Single GPU (`n_stages=1, n_workers=1`)
* Data Parallel (`n_stages=1, n_workers=n`)
* Pipeline Parallel (`n_stages=n, n_workers=n`)
* SWARM (`n_stages=m, n_workers=n`)

The experiments are run on 4xRTX 3090 on the [Prime Intellect Compute](https://api.primeintellect.ai) platform. We train GPT-2 (124M) for five steps on WikiText 2 (17.8M tokens)
with a batch size of ~0.5M tokens (512x1024). We look at the average and peak throughput.

View the experiment: [W&B](https://wandb.ai/mikasenghaas/swarm/table?nw=sx5y41kkz1q)

```python
# Load runs
TAG = "Benchmark"
assert False, "❌ This section uses old code that is not compatible anymore."
RUNS = [r for r in ALL_RUNS if TAG in r.tags and r.state == "finished"]

print(f"✅ Loaded {len(RUNS)} runs for experiment tagged with `{TAG}`")
```

```python
# Get config and summary
runs_config, runs_summary = {}, []
for run in tqdm(RUNS):
    _, group, name, config, summary = get_metadata(run)
    runs_config[(group, name)] = config
    runs_summary.append({"group": group, "name": name, **summary})

runs_summary = pd.DataFrame(runs_summary).set_index(["group", "name"])
```

```python
# Construct performance dataframe with only master nodes
performance = runs_summary.dropna().copy()

# Add varying constants
performance["num_nodes"] = performance.index.map(lambda x: runs_summary.groupby("group").size().loc[x[0]])
performance["num_stages"] = performance.index.map(lambda x: runs_config[x]["swarm"]["num_stages"])
performance["micro_batch_size"] = performance.index.map(lambda x: runs_config[x]["train"]["micro_batch_size"])
performance["max_micro_batches"] = performance.index.map(lambda x: runs_config[x]["train"].get("max_micro_batches", None))

# Drop name from index
performance = performance.reset_index().drop(columns=["name"]).set_index("group")
performance = performance.dropna()
performance["max_micro_batches"] = performance.max_micro_batches.astype(int)

# Add flag for type of parallelism
def get_parallelism(row: pd.Series):
    if row.num_nodes == 1: return "Single GPU"
    if row.num_stages == 1: return "Data Parallel"
    if row.num_stages == row.num_nodes: return "Pipeline Parallel"
    return "SWARM"
performance["parallelism"] = performance.apply(lambda row: get_parallelism(row), axis=1)

# Choose relevant columns
cols = ["parallelism", "num_nodes", "num_stages", "micro_batch_size", "max_micro_batches", "train/throughput/current"]
performance = performance[cols]

performance.head()
```

```python
single_gpu = performance[performance.parallelism == "Single GPU"]
dp = performance[performance.parallelism == "Data Parallel"]
pp = performance[performance.parallelism == "Pipeline Parallel"]
swarm = performance[performance.parallelism == "SWARM"]
```

### Comparison

```python
# Best configuration per parallelism type
peak_performance = performance.groupby("parallelism")\
    .apply(lambda x: x.loc[x["train/throughput/current"].idxmax()], include_groups=False)\
    .sort_values("train/throughput/current", ascending=False)

peak_performance
```

We get the best performance with traditional data parallelism, leading to an almost linear scaling of throughput with the number of nodes. Pipeline parallelism is more complex, as it can only process a maximum number of micro batches in parallel before halting and waiting for gradients to be computed. SWARM is in the middle of the two, combining data parallelism with pipeline parallelism.
